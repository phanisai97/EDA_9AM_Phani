```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns

filepath1=r"/Users/phani/Documents/Sai_Files/02Course_Files/07_Naresh_Data_Science/03FilesgivenbyOmkarSir/Data_Files/Visadataset.csv"
visa_df=pd.read_csv(filepath1)
visa_df

cat=visa_df.select_dtypes(include='object').columns.tolist()
num=visa_df.select_dtypes(exclude='object').columns.tolist()
```

```python
data = visa_df['prevailing_wage']
mean_data = data.mean()
std = data.std()
nr = data-mean_data
z_wage=nr/std
visa_df['z_wage']=z_wage
visa_df
```

```python
wage_data = visa_df['prevailing_wage']
plt.figure(figsize=(8,4))
plt.suptitle('Raw data vs Z-data')
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(z_wage)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(z_wage,vert=False)
plt.show()
```

# reshape can be applicable on numpyarray and dataframe but not on series

```python
# Step 1: import method
# Step 2: Save the method
# Step 3: Apply fit transform
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
wage_ss=scalar.fit_transform(visa_df['prevailing_wage'].values.reshape(-1,1))
# dir(StandardScalar)

# use [[]] to convert series into dataframe

visa_df['wage_ss']=wage_ss
visa_df['wage_ss']
```

# MinMaxScalar

```python
# Step-1: Read the data
# Step-2: Calc min
# Step-3: Calc max
# Step-4: Nr=step1-step2
# Step-5: Dr=step3-step2
# Step-6: Nr/Dr
# Step-7: visa_df['wage_min_max']=Nr/Dr
wage_data=visa_df['prevailing_wage']
xmin=wage_data.min()
xmax=wage_data.max()
Nr=wage_data-xmin
Dr=xmax-xmin
visa_df['wage_min_max']=Nr/Dr
visa_df['wage_min_max']
```

# Using sklearn MinMaxScalar

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
wage_min_max_2=scalar.fit_transform(visa_df[['prevailing_wage']])
visa_df['wage_min_max_skleanr']=wage_min_max_2
visa_df['wage_min_max_skleanr']
```

```python
max(visa_df['wage_min_max_skleanr']),min(visa_df['wage_min_max_skleanr'])
```
